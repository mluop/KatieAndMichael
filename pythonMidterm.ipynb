## Narrow down to definite labels

```python
def final(row):
    if row['status'] in ['Fully Paid']:
        val = 1
    elif row['status'] in ['Charged Off','Default']:
        val = 0
    else:
        val = -1
    return val

s['Final'] = s.apply(final, axis=1)
```

```python
df_filter.tail()
```

```python
df_filter.loc[df_filter['emp_length'] == 'n/a'].shape
```

```python
df_filter.replace('n/a', np.nan,inplace=True)
df_filter.emp_length.fillna(value=0,inplace=True)

df_filter['emp_length'].replace(to_replace='[^0-9]+', value='', inplace=True, regex=True)
df_filter['emp_length'] = df_filter['emp_length'].astype(int)
```

## No more null values

## LOGISTIC REGRESSION

## With L1

```python
classifierL1 = LogisticRegression(penalty='l1', random_state=0)
classifierL1.fit(X_train,y_train)
```

```python
y_pred_tr = classifierL1.predict(X_train)
from sklearn.metrics import confusion_matrix
confusion_matrix_tr = confusion_matrix(y_train,y_pred_tr)
print(confusion_matrix_tr)
```

```python
y_pred = classifierL1.predict(X_test)
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test,y_pred)
print(confusion_matrix)
```

```python
# left diag correct
# right diag incorrect
```

```python
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(classifierL1.score(X_test, y_test)))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

## With L2

```python
classifierL2 = LogisticRegression(penalty='l2', random_state=0)
classifierL2.fit(X_train,y_train)
```

```python
y_pred_tr = classifierL2.predict(X_train)
from sklearn.metrics import confusion_matrix
confusion_matrix_tr = confusion_matrix(y_train,y_pred_tr)
print(confusion_matrix_tr)
```

```python
y_pred = classifierL2.predict(X_test)
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test,y_pred)
print(confusion_matrix)
```
